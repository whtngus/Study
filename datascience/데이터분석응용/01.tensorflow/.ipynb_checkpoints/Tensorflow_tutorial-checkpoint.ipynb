### gpu cpu performance 비교

- Cpu 와 Gpu 성능 비교 하기!

```python
import time

def measuere(x,steps):
    with tf.Session() as sess:
        # 초기화시 gpu 속도가 느려지는 시간을 제외하기 위해서 
        sess.run(tf.matmul(x,x))
        start = time.time()
        for i in range(steps):
            x = tf.matmul(x,x)
        _ = sess.run(x)
        end = time.time()
        sess.close()
    return end - start
```

```python
# 장비 선택및 비교하기
shape = (10000,10000)
steps = 2
with tf.device("/cpu:0"):
    print("CPU : {} secs".format(measuere(tf.random_normal(shape),steps)))

if tf.test.is_gpu_available():
    with tf.device("/gpu:0"):
        print("GPU : {} secs".format(measuere(tf.random_normal(shape),steps)))
else:
    print("GPU not found")
```

### Computing gradients
- 미분 해보기

```python
x =  tf.constant(3.0)
init = tf.global_variables_initializer()
with tf.GradientTape() as g:
    g.watch(x)
    y = x*x

dy_dx = g.gradient(y,x)

with tf.Session() as sess:
    print(sess.run(dy_dx))
```

- 복잡한 연산에서 미분 해보기

```python
x =  tf.constant(3.0)
init = tf.global_variables_initializer()
# 기록
with tf.GradientTape() as g:
    g.watch(x)
    y = 1
    for i in range(3):
        if i < 2:
            y *= x
        else:
            y += x
        # x^2 + x

dy_dx = g.gradient(y,x)

with tf.Session() as sess:
    print(sess.run(dy_dx))
```

```python
x =  tf.constant(3.0)
init = tf.global_variables_initializer()
# 기록
with tf.GradientTape() as g:
    g.watch(x)
    y = 1
    with tf.GradientTape() as gg:
        gg.watch(x)
        y = x*x
    dy_dx = gg.gradient(y,x)
dy_dx2 = g.gradient(dy_dx,x)


with tf.Session() as sess:
    print(sess.run(dy_dx))
    print(sess.run(dy_dx2))
```

## 간단한 예제 진행해 보기

```python
# matplotlib 플로팅 도구를 
# %matplotlib.inline
import matplotlib.pyplot as plt 
import numpy as np 
import tensorflow as tf 
```

```python
np_x = np.array([0.0,1.0,1.0])
np_y = np.array([0.0,1.0,2.0])

# plt.plot(np_x,np_y,"o")
# plt.show()
```

```python
x = tf.placeholder(tf.float32)
a = tf.Variable(0.0)
b = tf.Variable(0.0)
y = a*x + b
```

```python
y_label = tf.placeholder(tf.float32)
# loss 함수 정의
loss =  tf.reduce_sum(tf.square(y - y_label))
```

- 학습을 하지 않은상태에서의 loss 를 계산해보기 


```python
# 시각화 먼저 
xs = np.arange(-0.1,1.1,0.1)
ys = 0*xs + 0
plt.plot(np_x,np_y,'o',xs,ys,'r')
plt.show()
```

```python
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    print(sess.run(loss,feed_dict={x:np_x,y_label:np_y}))
```

- loss 줄이기 위해 optimzer 를 정의하자

```python
# optimizer 선언
optimizer = tf.train.GradientDescentOptimizer(0.001)
# loss 를 optimizer 대상으로 선언하기
train_op = optimizer.minimize(loss)
```

```python
# 이제 train_op이 될때마다 weight를 update하게 됨
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for step in range(5000):
        _, loss_value, a_value, b_value = sess.run([train_op,loss,a,b], feed_dict={x:np_x,y_label:np_y})
        print("step : {} , loss : {} , a : {} , b : {}".format(step, loss_value,a_value, b_value))
```

- 학습한 값을 출력해보기

```python
xs = np.arange(-0.1,1.1,0.1)
ys = a_value*xs + b_value
plt.plot(np_x,np_y,"o",xs,ys,"r")
plt.show()
```

- 전체 정리및 다시 해보기

```python
# model
x = tf.placeholder(tf.float32)
a = tf.Variable(0.0)
b = tf.Variable(0.0)
c = tf.Variable(0.0)
y = a*x*x + b*x + c

# loss
y_label = tf.placeholder(tf.float32)
loss = tf.reduce_sum(tf.square(y-y_label))

# optimizer
optimizer = tf.train.GradientDescentOptimizer(0.01)
train_op = optimizer.minimize(loss)

# training
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for step in range(1000):
        _, loss_value, a_value, b_value, c_value =sess.run([train_op,loss,a,b,c], feed_dict={x:np_x,y_label:np_y})
```

```python
xs = np.arange(-0.1,1.1,0.1)
ys = a_value*xs*xs + b_value*xs + c_value
plt.plot(np_x,np_y,"o",xs,ys,"r")
plt.show()
```
